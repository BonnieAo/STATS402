```python
import pandas as pd
###show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
###Read the csv file
df=pd.read_csv('/content/uni.csv')
df.head()
```

```python
###Select the valid time interval
df = df[df['time']>'2020-09-17']
df = df[df['time']<'2021-04-25']
```

```python
df.head()
```

```python
df.info()
```

```python
#Price Rolling Average (30 days)
import matplotlib.pyplot as plt
df['SMA_30'] = df['PriceUSD'].rolling(window = 30).mean()
df[['PriceUSD','SMA_30']].plot(subplots=False, figsize=(12,6), grid=True, title='Uni Price (USD)')
plt.show()
```

```python
#Price Rolling Average (30 days) - Date
import plotly.express as px
fig_1 = px.line(df, x='time', y='PriceUSD')
fig_1.add_scatter(x=df['time'], y=df['SMA_30'],mode='lines') 
fig_1.update(layout=dict(xaxis=dict(title="Time", showline=True),
                       yaxis=dict(title="Transaction Value", showline=True)))
fig_1.show()
```

```python
#SplyFF time series plot
import matplotlib.pyplot as plt
df[['SplyFF']].plot(subplots=False, figsize=(12,6), grid=True, title='Supply')
plt.show()
```

```python
#TxTfrValUSD time series plot
import plotly.express as px
fig_1 = px.line(df, x='time', y='TxTfrValUSD') 
fig_1.update(layout=dict(xaxis=dict(title="Time", showline=True),
                       yaxis=dict(title="Transaction Value", showline=True)))
fig_1.show()
```

```python
#TxTfrValMeanUSD time series plot
fig_4 = px.line(df, x='time', y='TxTfrValMeanUSD')
fig_4.update(layout=dict(xaxis=dict(title="Time", showline=True),
                       yaxis=dict(title="Transaction Value Mean", showline=True)))
fig_4.show()
```

```python
#TxTfrValUSD Box Plot - To find outliers
df['box_plot_x'] = 1
fig_3 = px.box(df, x='box_plot_x', y='TxTfrValUSD', notched=True)
ax = fig_3.add_scatter(x=df['time'], y=df['SMA_30'],mode='lines') 
fig_3.update(layout=dict(xaxis=dict(title="TxTfrValUSD", showline=True),
                       yaxis=dict(title="Value", showline=True)))
fig_3.show()
```

```python
###Dealing with the outliers
df_scale = df.copy()
mu =df['TxTfrValUSD'].mean() 
sigma =df['TxTfrValUSD'].std() 
df = df[df['TxTfrValUSD']>mu+3*sigma]
```

```python
# Drop the outliers
df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]] = df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]].mean()
df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]] = df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]].mean()
df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]] = df_scale['TxTfrValMeanUSD'][df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]].mean()

df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]] = df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-07')].index.tolist()[0]].mean()
df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]] = df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-04')].index.tolist()[0]].mean()
df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]] = df_scale['TxTfrValUSD'][df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]-10:df_scale[(df_scale['time']=='2021-03-05')].index.tolist()[0]].mean()
```

```python
#Draw the correlation heatmap of the main features
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,12))
plt.title('Pearson Correlation of Features',y=1.05,size=15)
sns.heatmap(UNI_info.astype(float).corr(),linewidths=0.1,vmax=1.0,
            square=True,linecolor='white',annot=True)
plt.xticks(rotation=90)  
plt.yticks(rotation=360)
plt.show()
```
